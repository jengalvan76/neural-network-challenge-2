## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
attrition_df.shape
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
# Create X_df using your selected columns
X_df = attrition_df[['Education',
                     'Age',
                     'DistanceFromHome',
                     'JobSatisfaction',
                     'OverTime',
                     'StockOptionLevel',
                     'WorkLifeBalance',
                     'YearsAtCompany',
                     'YearsSinceLastPromotion',
                     'NumCompaniesWorked']]
X_df.head()

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
X_train.head()
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train["OverTime"].value_counts()
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
from sklearn.preprocessing import OneHotEncoder
encode = OneHotEncoder(sparse_output=False,drop="first",handle_unknown="ignore")
X_train["OverTime"] = encode.fit_transform(np.array(X_train["OverTime"]).reshape(-1,1))
X_test["OverTime"] = encode.fit_transform(np.array(X_test["OverTime"]).reshape(-1,1))
```

```python
X_train.dtypes
```

```python
# Create a StandardScaler
X_scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

X_train_scaled
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Fit the encoder to the training data
department_encode = OneHotEncoder(sparse_output=False,drop= None,handle_unknown="ignore").fit(np.array(y_train["Department"]).reshape(-1,1))
department_encode.categories_

# Create two new variables by applying the encoder
# to the training and testing data
department_y_train = department_encode.transform(np.array(y_train["Department"]).reshape(-1,1))
department_y_test = department_encode.transform(np.array(y_test["Department"]).reshape(-1,1))

department_y_train
```

```python
department_y_train.shape
```

```python
# Create a OneHotEncoder for the Attrition column
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Fit the encoder to the training data
attrition_encode = OneHotEncoder(sparse_output=False,drop= None,handle_unknown="ignore").fit(np.array(y_train["Attrition"]).reshape(-1,1))
attrition_encode.categories_

# Create two new variables by applying the encoder
# to the training and testing data
attrition_y_train = attrition_encode.transform(np.array(y_train["Attrition"]).reshape(-1,1))
attrition_y_test = attrition_encode.transform(np.array(y_test["Attrition"]).reshape(-1,1))

attrition_y_train
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_columns = X_train.shape[1]
num_columns

# Create the input layer
input_layer = layers.Input(shape=(num_columns,), name = "input_features")

# Create at least two shared layers
shared_layer_1 = layers.Dense(units=64, activation="relu", name = "shared_layer_1")(input_layer)
shared_layer_2 = layers.Dense(units=128, activation="relu", name = "shared_layer_2")(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(units=32, activation="sigmoid", name = "department_hidden")(shared_layer_2)


# Create the output layer
department_output = layers.Dense(units=3, activation="sigmoid", name = "department_output")(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden  = layers.Dense(units=32, activation="sigmoid", name = "attrition_hidden")(shared_layer_2)

# Create the output layer
attrition_output  = layers.Dense(units=2, activation="sigmoid", name = "attrition_output")(attrition_hidden)
```

```python
# Create the model
#model = Model(inputs=input_layer, outputs=[department_output, attrition_output])
model = Model(inputs=[department_hidden, attrition_hidden], outputs=[department_output, attrition_output])
# Compile the model
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics={"department_output": "accuracy", "attrition_output": "accuracy"})

# Summarize the model
model.summary()
```
